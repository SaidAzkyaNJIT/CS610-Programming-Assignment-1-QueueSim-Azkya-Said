```python
#IMPORTS AND CLASSES
import random as r
#An arrival function for a new customer arriving once every n minutes:
def arrival(n):
    x = r.random()
    if x < 1/n:
        x = 1
    else:
        x = 0
    return x
#A service time function with average service time of s:
def servicet(s):
    n = (2*s)
    y = r.random()
    x = (n-1)*y+1
    return int(round(x,2))
#customer, queue, and servicestation classes
class C:
    def __init__(self,cid,arrivet,servet):
        self.cid=cid
        self.arrivet=arrivet
        self.servet=servet
class Q:
    def __init__(self):
        self.q = []
    def empty(self):
        return not self.q
    def enq(self, x):
        self.q.append(x)
    def deq(self):
        if not self.empty():
            return self.q.pop(0)
    def size(self):
        return len(self.q)
class SS:
    def __init__(self,sid,cserved,remt):
        self.sid = int(sid)
        self.cserved = 0
        self.remt = 0
```

```python
#SINGLE QUEUE SIMULATION
def sim1(d,ar,s):
    t = 0
    cidcounter = 0
    qu = Q()
    ss1 = SS(1,0,0)
    ss2 = SS(2,0,0)
    ss3 = SS(3,0,0)
    ssl = [ss1,ss2,ss3]

    maxq = []
    ss1use = []
    ss2use = []
    ss3use = []
    servrate = []
    waiting_times = []

    while t < d or not qu.empty():
    
        if t < d and arrival(ar) == 1:
            #print(f'\nNew customer:')
            c = C(cidcounter,t,servicet(s))
            qu.enq(c)
            cidcounter += 1
            #print(f'Queue size: {qu.size()}')
            servrate.append(c.servet)
            maxq.append(qu.size())
            #print(f'Customer service time: {c.servet}') 

        if not qu.empty():
            for ss in ssl:
                if ss.remt == 0 and not qu.empty():
                    c = qu.deq()
                    ss.cserved += 1
                    ss.remt = c.servet   
                    waiting_time = t - c.arrivet
                    waiting_times.append(waiting_time)
                    #print(f'Customer queued to station {ss.sid}')
        
        for i in ssl:
            if i.remt > 0:
                i.remt -= 1

        t += 1

        #print(f'Station {ss.sid} remt: {ss.remt}')
        ss1use.append(ss1.remt)
        ss2use.append(ss2.remt)
        ss3use.append(ss3.remt)
        #print(ss1use)
        #print(ss2use)
        #print(ss3use)
    
    if qu.empty():
        print(f'Duration: {t} mins')
        print(f'{cidcounter} customers')
        print(f'Approximately one customer every {ar} minutes')
        print(f'\nStation 1 served: {ss1.cserved}')
        print(f'Station 2 served: {ss2.cserved}')
        print(f'Station 3 served: {ss3.cserved}\n')
        print(f'Max q size: {max(maxq)}')
        ss1usedp = round(1-(sum(1 for x in ss1use if x == 0)/len(ss1use)),2)
        print(f'Station 1 Rate of Occupancy: {ss1usedp}')
        ss2usedp = round(1-(sum(1 for x in ss2use if x == 0)/len(ss2use)),2)
        print(f'Station 2 Rate of Occupancy: {ss2usedp}')
        ss3usedp = round(1-(sum(1 for x in ss3use if x == 0)/len(ss3use)),2)
        print(f'Station 3 Rate of Occupancy: {ss3usedp}\n')
        print(f'Avg service time: {round(sum(servrate)/len(servrate),1)} mins')
        avg_wait = sum(waiting_times) / len(waiting_times)
        max_wait = max(waiting_times)
        print(f'Average Waiting Time: {round(avg_wait, 1)} mins')
        print(f'Maximum Waiting Time: {max_wait} mins')
```
